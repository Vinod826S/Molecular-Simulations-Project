```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.patches as mpatches


num_methanes = 4  # Number of methane molecules
a = 5  # Lattice constant
bond_length = 1.09  # Bond length in Angstrom
bond_angle = 109.5  # Bond angle in degrees

bond_angle_rad = np.radians(bond_angle)

# Create lattice positions
x = np.arange(0, num_methanes) * a
y = np.arange(0, num_methanes) * a
z = np.arange(0, num_methanes) * a
X, Y, Z = np.meshgrid(x, y, z)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for i in range(num_methanes):
    for j in range(num_methanes):
        for k in range(num_methanes):
            carbon_x = X[i,j,k]
            carbon_y = Y[i,j,k]
            carbon_z = Z[i,j,k]
            ax.scatter(carbon_x, carbon_y, carbon_z, color='gray')

            tetrahedral_positions = [
                (bond_length, bond_length, bond_length),
                (-bond_length, -bond_length, bond_length),
                (-bond_length, bond_length, -bond_length),
                (bond_length, -bond_length, -bond_length)
            ]

            for hx, hy, hz in tetrahedral_positions:
                ax.scatter(carbon_x + hx, carbon_y + hy, carbon_z + hz, color='red')

                ax.plot([carbon_x, carbon_x + hx], [carbon_y, carbon_y + hy], [carbon_z, carbon_z + hz], color='black', alpha=0.5, linestyle='--')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

carbon_patch = mpatches.Patch(color='gray', label='Carbon Atom')
hydrogen_patch = mpatches.Patch(color='red', label='Hydrogen Atom')

ax.legend(handles=[carbon_patch, hydrogen_patch], fontsize='small')

ax.set_title('Methane Structure in Lattice')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.patches as mpatches


num_ethanes = 4  # Number of ethane molecules
a = 5  # Lattice constant
cc_bond_length = 1.54  # C-C bond length in Angstrom
ch_bond_length = 1.1  # C-H bond length in Angstrom
cc_bond_angle = 111.2  # C-C bond angle in degrees
ch_bond_angle = 109.5  # C-H bond angle in degrees

cc_bond_angle_rad = np.radians(cc_bond_angle)
ch_bond_angle_rad = np.radians(ch_bond_angle)

x = np.arange(0, num_ethanes) * a
y = np.arange(0, num_ethanes) * a
z = np.arange(0, num_ethanes) * a
X, Y, Z = np.meshgrid(x, y, z)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for i in range(num_ethanes):
    for j in range(num_ethanes):
        for k in range(num_ethanes):
            carbon1_x = X[i,j,k] + 0.5 * a
            carbon1_y = Y[i,j,k] + 0.5 * a
            carbon1_z = Z[i,j,k] + 0.5 * a

            carbon2_x = X[i,j,k] - 0.5 * a
            carbon2_y = Y[i,j,k] - 0.5 * a
            carbon2_z = Z[i,j,k] - 0.5 * a

            ax.scatter(carbon1_x, carbon1_y, carbon1_z, color='gray')
            ax.scatter(carbon2_x, carbon2_y, carbon2_z, color='gray')

            ax.plot([carbon1_x, carbon2_x], [carbon1_y, carbon2_y], [carbon1_z, carbon2_z], color='black', alpha=0.7)

            tetrahedral_positions_c1 = [
                (cc_bond_length, -cc_bond_length * np.cos(cc_bond_angle_rad), 0),  # H1 (behind)
                (cc_bond_length * np.cos(cc_bond_angle_rad), 0, cc_bond_length),  # H2 (right)
                (0, cc_bond_length * np.cos(cc_bond_angle_rad), cc_bond_length),  # H3 (left)
            ]

            for hx, hy, hz in tetrahedral_positions_c1:
                ax.scatter(carbon1_x + hx, carbon1_y + hy, carbon1_z + hz, color='red')
                ax.plot([carbon1_x, carbon1_x + hx], [carbon1_y, carbon1_y + hy], [carbon1_z, carbon1_z + hz], color='black', alpha=0.5, linestyle='--')

            tetrahedral_positions_c2 = [
                (-cc_bond_length, cc_bond_length * np.cos(cc_bond_angle_rad), 0),  # H1 (front)
                (-cc_bond_length * np.cos(cc_bond_angle_rad), 0, -cc_bond_length),  # H2 (right)
                (0, -cc_bond_length * np.cos(cc_bond_angle_rad), -cc_bond_length),  # H3 (left)
            ]

            for hx, hy, hz in tetrahedral_positions_c2:
                ax.scatter(carbon2_x + hx, carbon2_y + hy, carbon2_z + hz, color='red')
                ax.plot([carbon2_x, carbon2_x + hx], [carbon2_y, carbon2_y + hy], [carbon2_z, carbon2_z + hz], color='black', alpha=0.5, linestyle='--')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

carbon_patch = mpatches.Patch(color='gray', label='Carbon Atom')
hydrogen_patch = mpatches.Patch(color='red', label='Hydrogen Atom')

ax.legend(handles=[carbon_patch, hydrogen_patch], fontsize='small')

ax.set_title('Ethane Structure in Lattice')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
import matplotlib.patches as mpatches

num_methanes = 3  # Number of methane molecules
a = 5  # Lattice constant
bond_length = 1.09  # Bond length in Angstrom
bond_angle = 109.5  # Bond angle in degrees
num_frames = 150 # Number of frames in the animation

# Convert bond angle to radians
bond_angle_rad = np.radians(bond_angle)

# Create lattice positions
x = np.arange(0, num_methanes) * a
y = np.arange(0, num_methanes) * a
z = np.arange(0, num_methanes) * a
X, Y, Z = np.meshgrid(x, y, z)

# Function to generate random velocities
def generate_random_velocity():
    return np.random.normal(0, 0.2, 3)

# Function to update positions
def update_positions(frame):
    ax.clear()

    for i in range(num_methanes):
        for j in range(num_methanes):
            for k in range(num_methanes):
                carbon_x = X[i,j,k] + frame * 0.01 * generate_random_velocity()[0]
                carbon_y = Y[i,j,k] + frame * 0.01 * generate_random_velocity()[1]
                carbon_z = Z[i,j,k] + frame * 0.01 * generate_random_velocity()[2]
                ax.scatter(carbon_x, carbon_y, carbon_z, color='gray', s=100)

                tetrahedral_positions = [
                    (bond_length, bond_length, bond_length),
                    (-bond_length, -bond_length, bond_length),
                    (-bond_length, bond_length, -bond_length),
                    (bond_length, -bond_length, -bond_length)
                ]

                for hx, hy, hz in tetrahedral_positions:
                    ax.scatter(carbon_x + hx, carbon_y + hy, carbon_z + hz, color='red', s=50)

                    ax.plot([carbon_x, carbon_x + hx], [carbon_y, carbon_y + hy], [carbon_z, carbon_z + hz],
                            color='black', alpha=0.5, linestyle='--')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Methane Molecular Dynamics Simulation')

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ani = FuncAnimation(fig, update_positions, frames=num_frames, interval=33)

ani.save('Methane Simulation.mp4', writer='ffmpeg')
```
